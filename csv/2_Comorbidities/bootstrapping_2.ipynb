```python
import pandas as pd

dfc = pd.read_csv('comorb_detallades.csv')

# "Resample" para llegar a 100,000 filas
# replace=True permite que un paciente se repita
dfc_100 = dfc.sample(n=100000, replace=True, random_state=42)

dfc_100.to_csv('comorb_detallades_100k.csv', index=False)
```

```python
import pandas as pd
import numpy as np

# Definir el volum total
n_total = 100000

# 1. Definir les probabilitats (Freqüències)
smoking_labels = ['Ever smoker', 'Never smoker']
smoking_probs = [0.4151, 0.5849]

# 3. GENERACIÓ DE LA TAULA 2
data_taula2 = {
    'id': range(1, n_total + 1), # Mateix ID que la Taula 1
    'smoking_status': np.random.choice(smoking_labels, size=n_total, p=smoking_probs)
}

df_taula2 = pd.DataFrame(data_taula2)

# 4. GUARDAR EL FITXER
df_taula2.to_csv('comorbidities_100k.csv', index=False)
```
