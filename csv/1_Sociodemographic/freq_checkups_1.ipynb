## Age

```python
df2=pd.read_csv('MOCK_DATA.csv')
df2.age.value_counts(normalize=True)*100
```

```python
# 1. Definir els punts de tall (límits dels rangs)
# Els talls han de ser: 0-14, 15-19, 20-34, 35-64, i >65
# Posem 120 com a límit superior per a la gent gran
talls = [0, 14, 19, 34, 64, 120]
etiquetes = ['0-14', '15-19', '20-34', '35-64', '>65']

# 2. Agrupar les edats en aquests rangs temporalment
# Fem servir include_lowest=True perquè inclogui el 0
edat_per_rangs = pd.cut(df2['age'], bins=talls, labels=etiquetes, include_lowest=True)

# 3. Calcular i imprimir la freqüència
print("--- Freqüència per Rangs d'Edat ---")
counts = edat_per_rangs.value_counts(normalize=True).sort_index() * 100

print(counts.round(2).astype(str) + '%')
```

```python
# 1. Carrego el fitxer on vull fer el canvi
df = pd.read_csv('MOCK_DATA_N_age_mod.csv')

# 2. Substituïm la columna 'age' per la de l'altre dataframe (df2)
# IMPORTANT: Assegura't que els dos fitxers tinguin el mateix nombre de files
df['age'] = df2['age']

# 3. Guardem el resultat sobreescrivint el fitxer o creant-ne un de nou
# Fem servir index=False per no afegir una columna extra de números a l'esquerra
df.to_csv('MOCK_DATA_N_age_mod.csv', index=False)
df.head()
```

## Smoking Status

```python
df.smoking_status.value_counts(normalize=True)*100
```

```python
# 1. Definir les probabilitats segons la taula 1 del document PADRIS-PRESTO
categories = ['Ever smoker', 'Never smoker']
weights = [0.4151, 0.5849] # 41.51% i 58.49%

# 2. Aplicar la distribució a la columna smoking_status
# Utilitzem len(df) per assegurar que omplim totes les files del teu fitxer
df['smoking_status'] = np.random.choice(categories, size=len(df), p=weights)

# 3. Aplicar cobertura del 54,3%
num_buits = int(len(df) * 0.457)
indexs_buits = np.random.choice(df.index, size=num_buits, replace=False)
df.loc[indexs_buits, 'smoking_status'] = np.nan

# 4. Guardar el fitxer amb els canvis
df.to_csv('MOCK_DATA_N_age_mod.csv', index=False)
print(df['smoking_status'].value_counts(normalize=True, dropna=False) * 100)
```
