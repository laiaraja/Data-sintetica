```python
import pandas as pd
import os

# 1. Configura la carpeta on tens els fitxers pesats
carpeta_entrada = r'C:\Users\LRAJAG\Desktop\DATOS_PADRIS'
carpeta_sortida = r'C:\Users\LRAJAG\Desktop\DATOS_PADRIS_nets'

# Creem la carpeta de sortida si no existeix
if not os.path.exists(carpeta_sortida):
    os.makedirs(carpeta_sortida)

# 2. Bucle per recórrer tots els fitxers de la carpeta
for fitxer in os.listdir(carpeta_entrada):
    if fitxer.endswith(".csv"):
        print(f"🔄 Processant: {fitxer}...")
        
        path_complet_entrada = os.path.join(carpeta_entrada, fitxer)
        path_complet_sortida = os.path.join(carpeta_sortida, fitxer)
        
        # Llegim amb ~ i guardem amb coma (usant chunksize per si són enormes)
        chunk_size = 100000
        reader = pd.read_csv(path_complet_entrada, sep='~', chunksize=chunk_size, low_memory=False)
        
        for i, chunk in enumerate(reader):
            mode = 'w' if i == 0 else 'a'
            header = True if i == 0 else False
            chunk.to_csv(path_complet_sortida, mode=mode, index=False, sep=',', header=header)
            
        print(f"✅ Finalitzat: {fitxer}")

print("\n Tots els fitxers han estat convertits!")
```
