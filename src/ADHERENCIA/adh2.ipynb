# Adherance per years code

```python
# We assume dispensing happens in the middle of the month, so we set day to 15 for all records
def anymes_to_date(anymes):
    anymes = str(anymes)
    return pd.Timestamp(year=int(anymes[:4]), month=int(anymes[4:]), day=15)

df_23_dispensats['dispensing_date'] = df_23_dispensats['Any_Mes'].apply(anymes_to_date)
df_25_mhda['dispensing_date'] = df_25_mhda['Any_Mes'].apply(anymes_to_date)
```

```python
import pandas as pd
import numpy as np
from datetime import date

# Definition of the merging of the dispensing intervals, the calculation is days covered in a year and leftover days are carried over to the next interval if the patient refills early. 
def merge_intervals(intervals):
    """Carry over leftover days when patient refills early."""
    if not intervals:
        return []
    sorted_intervals = sorted(intervals, key=lambda x: x[0])
    result = [sorted_intervals[0]]
    
    for start, end in sorted_intervals[1:]:
        prev_end = result[-1][1]
        if start <= prev_end:
            # Refill before finishing — shift new interval forward
            leftover = (prev_end - start).days + 1
            new_end = end + pd.Timedelta(days=leftover)
            result.append((prev_end + pd.Timedelta(days=1), new_end))
        else:
            result.append((start, end))
    
    return result

# Calculate days covered in a year, accounting for treatment start and end of year
def days_covered_in_year(intervals, year, treatment_start):
    """Count days covered within a calendar year."""
    year_start = max(pd.Timestamp(year, 1, 1), treatment_start)
    year_end = pd.Timestamp(year, 12, 31)
    
    denominator = (year_end - year_start).days + 1
    
    covered = 0
    for start, end in intervals:
        overlap_start = max(start, year_start)
        overlap_end = min(end, year_end)
        if overlap_start <= overlap_end:
            covered += (overlap_end - overlap_start).days + 1
    
    return min(covered, denominator), denominator

# Main function to compute PDC per year for one patient, combining both ORAL and LAI intervals and calculating PDC separately for each formulation type as well as combined.
def compute_pdc(patient_df_23, patient_df_25, treatment_start):
    """Compute PDC per year for one patient."""
    results = []
    
    # Build coverage intervals for ORAL
    oral_intervals = []
    for _, row in patient_df_23[patient_df_23['formulation_type'] == 'ORAL'].iterrows():
        start = row['dispensing_date']
        end = start + pd.Timedelta(days=int(row['days_coverage']) - 1)
        oral_intervals.append((start, end))
    
    # Build coverage intervals for LAI (both tables)
    lai_intervals = []
    for _, row in patient_df_23[patient_df_23['formulation_type'] == 'LAI'].iterrows():
        start = row['dispensing_date']
        end = start + pd.Timedelta(days=row['days_coverage'] - 1)
        lai_intervals.append((start, end))
    for _, row in patient_df_25[patient_df_25['formulation_type'] == 'LAI'].iterrows():
        start = row['dispensing_date']
        end = start + pd.Timedelta(days=row['days_coverage'] - 1)
        lai_intervals.append((start, end))
    
    oral_merged = merge_intervals(oral_intervals)
    lai_merged = merge_intervals(lai_intervals)
    total_merged = merge_intervals(oral_intervals + lai_intervals)
    
    # Get all years from treatment start to 2019
    years = range(treatment_start.year, 2020)
    
    for year in years:
        oral_covered, denominator = days_covered_in_year(oral_merged, year, treatment_start)
        lai_covered, _ = days_covered_in_year(lai_merged, year, treatment_start)
        total_covered, _ = days_covered_in_year(total_merged, year, treatment_start)
        
        results.append({
            'year': year,
            'denominator': denominator,
            'PDC_oral': round(oral_covered / denominator, 4),
            'PDC_lai': round(lai_covered / denominator, 4),
            'PDC_total': round(total_covered / denominator, 4),
        })
    
    return pd.DataFrame(results)
```

```python
df_23_dispensats['days_coverage'] = df_23_dispensats.apply(
    lambda row: LAI_INTERVALS[row['DES_HIS_Producte']] 
    if row['formulation_type'] == 'LAI' 
    else (float(str(row['Nombre_DDD_Receptes_Dispensades']).replace(',', '.')) 
          if row['formulation_type'] == 'ORAL' else 0),
    axis=1
)

print(df_23_dispensats[['DES_HIS_Producte', 'formulation_type', 'days_coverage']].sample(5))
```

```python
all_results = []

patients = cohort['idCas'].unique()

for patient_id in patients:
    p23 = df_23_dispensats[df_23_dispensats['idCas'] == patient_id]
    p25 = df_25_mhda[df_25_mhda['idCas'] == patient_id]
    
    # Treatment start = earliest dispensing across both tables
    dates = []
    if not p23.empty:
        dates.append(p23['dispensing_date'].min())
    if not p25.empty:
        dates.append(p25['dispensing_date'].min())
    
    if not dates:
        continue
    
    treatment_start = min(dates)
    
    pdc_df = compute_pdc(p23, p25, treatment_start)
    pdc_df['idCas'] = patient_id
    all_results.append(pdc_df)

pdc_final = pd.concat(all_results, ignore_index=True)
pdc_final = pdc_final.sort_values(['idCas', 'year']).reset_index(drop=True)
print(pdc_final.head(20))
print(pdc_final.shape)
```

```python
# First oral dispensing per patient
first_oral = df_23_dispensats[df_23_dispensats['formulation_type'] == 'ORAL'].groupby('idCas')['dispensing_date'].min().reset_index()
first_oral.columns = ['idCas', 'first_oral_date']

# First LAI dispensing per patient (both tables)
lai_t23 = df_23_dispensats[df_23_dispensats['formulation_type'] == 'LAI'][['idCas', 'dispensing_date']]
lai_t25 = df_25_mhda[df_25_mhda['formulation_type'] == 'LAI'][['idCas', 'dispensing_date']]
first_lai = pd.concat([lai_t23, lai_t25]).groupby('idCas')['dispensing_date'].min().reset_index()
first_lai.columns = ['idCas', 'first_lai_date']

# Merge
patient_flags = first_oral.merge(first_lai, on='idCas', how='outer')

# Switcher: had oral before LAI
patient_flags['switcher'] = (
    patient_flags['first_oral_date'].notna() & 
    patient_flags['first_lai_date'].notna() & 
    (patient_flags['first_oral_date'] < patient_flags['first_lai_date'])
)

# LAI prevalent: any LAI ever
patient_flags['lai_ever'] = patient_flags['first_lai_date'].notna()

# Oral only: never had LAI
patient_flags['oral_only'] = patient_flags['first_lai_date'].isna()

print(patient_flags[['switcher', 'lai_ever', 'oral_only']].value_counts())
```

We have:
- 152,991 oral only (never on LAI)
- 10,912 switchers (oral before LAI)
- 5,580 LAI prevalent (on LAI but no prior oral — started directly on LAI)

```python
pdc_final = pdc_final.merge(patient_flags[['idCas', 'switcher', 'lai_ever', 'oral_only', 'first_lai_date']], on='idCas', how='left')

# Filter to 2013-2019
pdc_final = pdc_final[pdc_final['year'].between(2013, 2021)]

print(pdc_final.head())
```

```python
def assign_group(row):
    if row['switcher']:
        return 'Switcher'
    elif row['oral_only']:
        return 'Oral only'
    else:
        return 'LAI prevalent'

patient_flags['group'] = patient_flags.apply(assign_group, axis=1)

# Just merge the group label
pdc_final = pdc_final.merge(patient_flags[['idCas', 'group']], on='idCas', how='left')
pdc_final = pdc_final[pdc_final['year'].between(2013, 2021)]
```
