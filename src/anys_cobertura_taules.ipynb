### Reorder columns per years and visits count

```python
# Get all data columns (excluding id and idCas)
data_cols = [col for col in df_9_2011_2019.columns if col not in ["id", "idCas"]]

# Get unique years and variable prefixes
years = sorted(set(col.split("_")[-1] for col in data_cols))
prefixes = sorted(set("_".join(col.split("_")[:-1]) for col in data_cols))

# Reorder columns by year
ordered_cols = ["id", "idCas"]
for year in years:
    year_cols = [col for col in data_cols if col.endswith(year)]
    ordered_cols += year_cols
    # Add total column per year
    df_9_2011_2019[f"TOTAL_{year}"] = df_9_2011_2019[year_cols].sum(axis=1)
    ordered_cols.append(f"TOTAL_{year}")

df_9_2011_2019 = df_9_2011_2019[ordered_cols]
print(df_9_2011_2019.columns.tolist())
print(df_9_2011_2019.head())
```

```python
# Fill NaN with 0 and convert to int
for col in data_cols:
    df_9_2011_2019[col] = df_9_2011_2019[col].fillna(0).astype(int)
```

```python
# Fill NaN with 0 and convert to int for all columns except id and idCas
for col in df_9_2011_2019.columns:
    if col not in ["id", "idCas"]:
        df_9_2011_2019[col] = df_9_2011_2019[col].fillna(0).astype(int)
```

## 7. SMH Tables 12, 13, 14

```python
df_schiz = pd.read_csv(r"C:\Users\laiar\Desktop\Classe IDIBAPS\data_sintetica\Data-sintetica\Data-sintetica\src\csv merged\cohort_schizophrenia_sociodemografica.csv")['idCas'].unique()
df_12_epi_smh = pd.read_csv(r"C:\Users\laiar\Desktop\DATOS_PADRIS_nets\12_taula_smh_epi_cohort_grup_estudi_Final.csv")
df_13_dx_smh= pd.read_csv(r"C:\Users\laiar\Desktop\DATOS_PADRIS_nets\13_taula_smh_dx_cohort_grup_estudi_Final.csv")
df_14_proc_smh = pd.read_csv(r"C:\Users\laiar\Desktop\DATOS_PADRIS_nets\14_taula_smh_p_cohort_grup_estudi_Final.csv")

df_12_epi_smh_f = df_12_epi_smh[df_12_epi_smh['idCas'].isin(df_schiz)]
df_13_dx_smh_f = df_13_dx_smh[df_13_dx_smh['idCas'].isin(df_schiz)]
df_14_proc_smh_f = df_14_proc_smh[df_14_proc_smh['idCas'].isin(df_schiz)]
```

```python
# Taula 12 - només dat_ini
df_12_slim = df_12_epi_smh_f[['idCas', 'id_cmbd', 'dat_ini']]

# Taula 13 - desc_dx positions 0-3 in columns
df_13_wide = df_13_dx_smh_f[df_13_dx_smh_f['pos_dx'].isin([1,2,3,4])]\
    .pivot_table(index=['idCas', 'id_cmbd'], columns='pos_dx', values='desc_dx', aggfunc='first')\
    .reindex(columns=[1,2,3,4])\
    .reset_index()

df_13_wide.columns = ['idCas', 'id_cmbd', 'desc_dx_1', 'desc_dx_2', 'desc_dx_3', 'desc_dx_4']

# Afegir pos_dx (és simplement 1,2,3,4 si existeix valor, NaN si no)
for i in [1,2,3,4]:
    df_13_wide.insert(
        df_13_wide.columns.get_loc(f'desc_dx_{i}'),
        f'pos_dx_{i}',
        df_13_wide[f'desc_dx_{i}'].apply(lambda x: i if pd.notna(x) else None)
    )

# Taula 14 - només GZ3ZZZZ, la resta NaN
df_14_slim = df_14_proc_smh_f[df_14_proc_smh_f['cod_proc'] == 'GZ3ZZZZ'][['idCas', 'id_cmbd', 'cod_proc', 'desc_proc']]

# Merge
df_merged_smh = df_12_slim\
    .merge(df_13_wide, on=['idCas', 'id_cmbd'], how='left')\
    .merge(df_14_slim, on=['idCas', 'id_cmbd'], how='left')
```

## 8. URG Tables 16, 17, 18

```python
# Carregar taules
df_16_urg_epi = pd.read_csv(r"C:\Users\laiar\Desktop\DATOS_PADRIS_nets\16_cmbd_urg_episodis_final_Final.csv")
df_17_urg_dx = pd.read_csv(r"C:\Users\laiar\Desktop\DATOS_PADRIS_nets\17_cmbd_urg_diagnostics_cohort_grup_estudi_Final.csv")
df_18_urg_proc = pd.read_csv(r"C:\Users\laiar\Desktop\DATOS_PADRIS_nets\18_cmbd_urg_p_cohort_grup_estudi_Final.csv")

# Filtrar per cohort
df_16_f = df_16_urg_epi[df_16_urg_epi['idCas'].isin(df_schiz)]
df_17_f = df_17_urg_dx[df_17_urg_dx['idCas'].isin(df_schiz)]
df_18_f = df_18_urg_proc[df_18_urg_proc['idCas'].isin(df_schiz)]

# Taula 16 - columnes seleccionades
df_16_slim = df_16_f[['idCas', 'id_cmbd', 'dat_ini', 'dat_fi', 'cod_proc_ing', 'desc_proc_ing', 'cod_tip_up_desti', 'desc_tip_up_desti']]

# Taula 17 - pivot desc_dx posicions 1-4
df_17_wide = df_17_f[df_17_f['pos_dx'].isin([1,2,3,4])]\
    .pivot_table(index=['idCas', 'id_cmbd'], columns='pos_dx', values='desc_dx', aggfunc='first')\
    .reindex(columns=[1,2,3,4])\
    .reset_index()
df_17_wide.columns = ['idCas', 'id_cmbd', 'desc_dx_1', 'desc_dx_2', 'desc_dx_3', 'desc_dx_4']
for i in [1,2,3,4]:
    df_17_wide.insert(df_17_wide.columns.get_loc(f'desc_dx_{i}'), f'pos_dx_{i}',
        df_17_wide[f'desc_dx_{i}'].apply(lambda x: i if pd.notna(x) else None))

# Taula 18 - pivot desc_proc posicions 0-3
df_18_wide = df_18_f[df_18_f['pos_proc'].isin([0,1,2,3])]\
    .pivot_table(index=['idCas', 'id_cmbd'], columns='pos_proc', values='desc_proc', aggfunc='first')\
    .reindex(columns=[0,1,2,3])\
    .reset_index()
df_18_wide.columns = ['idCas', 'id_cmbd', 'desc_proc_0', 'desc_proc_1', 'desc_proc_2', 'desc_proc_3']
for i in [0,1,2,3]:
    df_18_wide.insert(df_18_wide.columns.get_loc(f'desc_proc_{i}'), f'pos_proc_{i}',
        df_18_wide[f'desc_proc_{i}'].apply(lambda x: i if pd.notna(x) else None))

# Merge
df_merged_urg = df_16_slim\
    .merge(df_17_wide, on=['idCas', 'id_cmbd'], how='left')\
    .merge(df_18_wide, on=['idCas', 'id_cmbd'], how='left')
```
